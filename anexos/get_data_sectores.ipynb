```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import sqlalchemy

db = sqlalchemy.create_engine("sqlite:///data.sqlite")

def get_sectors(country):
    url = f"https://es.tradingview.com/markets/stocks-{country}/sectorandindustry-sector/"

    driver = webdriver.Edge()
    driver.get(url)

    links = driver.find_elements(By.XPATH, "//td/a")
    sectors = []
    for a in links:
        sectors.append({
            "name": a.text,
            "href": a.get_property("href")
        })
    
    pd.DataFrame(sectors).to_sql(f"{country}_sectors", db, if_exists="replace")

get_sectors("chile")
get_sectors("colombia")
get_sectors("mexico")
get_sectors("peru")
```
